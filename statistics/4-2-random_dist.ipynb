```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first
import thinkstats2
import thinkplot
```

```python
import numpy as np
numbers=[]
numbers=np.random.random(1000)
```

```python
pmf=thinkstats2.Pmf(numbers)
thinkplot.Pdf(pmf)
```

#### Answer : 
The data is not normally distributed but equally distributed and will have equal number of occurances for each number.


```python
cdf=thinkstats2.Cdf(numbers,label="Number distribution")
thinkplot.Cdf(cdf)
thinkplot.Config(xlabel="Probability",ylabel="Number")
```

#### Answer :
The individual probability of each occurance is equal hence the cummulative probability of all occurances is still One. And hence the shape of the graph is linear in nature.
